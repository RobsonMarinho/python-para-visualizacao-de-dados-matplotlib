```python
#Importando biblioteca e acessando a classe pyplot
import matplotlib.pyplot as plt
```

```python
#Visualizar os comandos na linha (inline)
%matplotlib inline
```

```python
import numpy as np
```

```python
#Cria um espaço vetorial inserindo os 11 primeiros elemntos entre 0 e 5 
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
# Funcional 
plt.plot(x, y, 'r--')
plt.xlabel('Fixo X')
plt.ylabel('Fixo Y')
plt.title('Título')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
#Cria uma caixa vazia
fig = plt.figure()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 1.0, 1.0])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Título')
axes2.plot(y, x)
```

```python
fug, ax = plt.subplots() #Cria 02 instâncias
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g--')
ax.set_ylabel('y')
ax.set_title('Título')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
ax[0].plot(x, y, 'b')
ax[0].set_title('Titulo')

ax[1].plot(x, y, 'b')
ax[1].set_title('Titulo')
fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
#Gestão da figura
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12, 10))
axes.plot(x, y, 'r-', label='Dados 1')
axes.plot(y, x, 'g--', label='Dados 2')
axes.set_title('Título')
axes.legend(loc=0)
```
