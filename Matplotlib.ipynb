```python
#Importando biblioteca e acessando a classe pyplot
import matplotlib.pyplot as plt
```

```python
#Visualizar os comandos na linha (inline)
%matplotlib inline
```

```python
import numpy as np
```

```python
#Cria um espaço vetorial inserindo os 11 primeiros elemntos entre 0 e 5 
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
# Funcional 
plt.plot(x, y, 'r--')
plt.xlabel('Fixo X')
plt.ylabel('Fixo Y')
plt.title('Título')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
#Cria uma caixa vazia
fig = plt.figure()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 1.0, 1.0])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Título')
axes2.plot(y, x)
```
